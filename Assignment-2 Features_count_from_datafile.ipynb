```python
def partsspeech(line):
    tagger=nltk.pos_tag(nltk.word_tokenize(line))
    no_of_nouns=no_of_verbs=no_of_adjectives=no_of_interjunction=no_of_prepositions=no_of_conjunctions=0
    for value,tag in tagger:
        if tag=="NN":
            no_of_nouns+=1
        elif tag=="VBZ":
            no_of_verbs+=1
        elif tag=="JJ":
            no_of_adjectives+=1
        elif tag=="IN":
            no_of_interjunction+=1
        elif tag=="PRP":
            no_of_prepositions+=1
        elif tag=="CC":
            no_of_conjunctions+=1
    return([no_of_nouns,no_of_verbs,no_of_adjectives,no_of_interjunction,no_of_prepositions,no_of_conjunctions])
```

```python
def specialchars(line):
    words=nltk.word_tokenize(line)
    no_ofalpha=no_ofdigits=no_ofspaces=no_ofanum=0
    for i in words:
        if i.isalpha():
            no_ofalpha+=1
        elif i.isdigit():
            no_ofdigits+=1
        elif i.isspace():
            no_ofspaces+=1
        elif i.isalnum():
            no_ofanum+=1
    return([no_ofalpha,no_ofdigits,no_ofspaces,no_ofanum])
```
